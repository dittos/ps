```python
with open("../input/day3.txt") as fp:
    data = fp.read().splitlines()
```

```python
import re

numbers = []
for row in data:
    numbers.append(list(re.finditer("([0-9]+)", row)))

numbers
```

```python
part_numbers = set()

for y, row in enumerate(data):
    for m in re.finditer("[^.0-9\n]", row):
        x = m.start()
        for qx, qy in (
            (x - 1, y - 1),
            (x, y - 1),
            (x + 1, y - 1),

            (x - 1, y),
            (x + 1, y),

            (x - 1, y + 1),
            (x, y + 1),
            (x + 1, y + 1),
        ):
            try:
                numbers_qy = numbers[qy]
            except IndexError:
                continue

            for num in numbers_qy:
                if num.start() <= qx < num.end():
                    part_numbers.add(num)

sum(int(num.group()) for num in part_numbers)
        
```

```python
answer2 = 0

for y, row in enumerate(data):
    for m in re.finditer("[^.0-9\n]", row):
        part_numbers = set()
        x = m.start()
        for qx, qy in (
            (x - 1, y - 1),
            (x, y - 1),
            (x + 1, y - 1),

            (x - 1, y),
            (x + 1, y),

            (x - 1, y + 1),
            (x, y + 1),
            (x + 1, y + 1),
        ):
            try:
                numbers_qy = numbers[qy]
            except IndexError:
                continue

            for num in numbers_qy:
                if num.start() <= qx < num.end():
                    part_numbers.add(num)
        if m.group() == "*" and len(part_numbers) == 2:
            # gear
            a, b = list(part_numbers)
            gear_ratio = int(a.group()) * int(b.group())
            answer2 += gear_ratio

answer2
        
```
