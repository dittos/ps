```python
with open("../input/day2.txt") as fp:
    games = []
    for line in fp:
        # Game {id}: {r0} red, {b0} blue, {g0} green; {r1} red, {b1} blue, {g1} green; ...
        game_id, revealed_list = line.strip().split(": ")
        game_id = game_id.split(" ")[-1]
        revealed_list_parsed = []
        for s in revealed_list.split("; "):
            rgb = [0, 0, 0]
            for n_with_color in s.split(", "):
                n, color = n_with_color.split(" ")
                n = int(n)
                if color == "red":
                    rgb[0] = n
                elif color == "green":
                    rgb[1] = n
                elif color == "blue":
                    rgb[2] = n
                else:
                    raise ValueError(color)
            revealed_list_parsed.append(rgb)
        games.append((game_id, revealed_list_parsed))
```

```python
games
```

```python
answer = 0
for game_id, revealed_list in games:
    min_r, min_g, min_b = map(max, zip(*revealed_list))
    if min_r <= 12 and min_g <= 13 and min_b <= 14:
        answer += int(game_id)
```

```python
answer2 = 0
for game_id, revealed_list in games:
    min_r, min_g, min_b = map(max, zip(*revealed_list))
    power = min_r * min_g * min_b
    answer2 += power

answer2
```
